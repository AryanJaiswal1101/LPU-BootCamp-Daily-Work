```python
# Define a list
my_list = [1, 2, 3, 4, 5]

# Question: How do you access the first element of the list?
print("Question: How do you access the first element of the list?")
print("Answer:", my_list[0])

# Question: How do you access the last element of the list?
print("Question: How do you access the last element of the list?")
print("Answer:", my_list[-1])

# Question: How do you add an element to the end of the list?
print("Question: How do you add an element to the end of the list?")
my_list.append(6)
print("Answer:", my_list)

# Question: How do you remove an element from the list?
print("Question: How do you remove an element from the list?")
my_list.remove(3)
print("Answer:", my_list)

# Question: How do you check if an element is in the list?
print("Question: How do you check if an element is in the list?")
print("Answer:", 5 in my_list)

# Question: How do you find the length of the list?
print("Question: How do you find the length of the list?")
print("Answer:", len(my_list))

# Question: How do you create a copy of the list?
print("Question: How do you create a copy of the list?")
copied_list = my_list.copy()
print("Answer:", copied_list)

# Question: How do you clear all elements from the list?
print("Question: How do you clear all elements from the list?")
my_list.clear()
print("Answer:", my_list)
```

```python
# Define a list
my_list = [1, 2, 3, 4, 5]

# Question: How do you insert an element at a specific position in the list?
print("Question: How do you insert an element at a specific position in the list?")
my_list.insert(2, 10)  # Insert 10 at index 2
print("Answer:", my_list)

# Question: How do you count the occurrences of a specific element in the list?
print("Question: How do you count the occurrences of a specific element in the list?")
count_of_2 = my_list.count(2)
print("Answer:", count_of_2)

# Question: How do you reverse the order of elements in the list?
print("Question: How do you reverse the order of elements in the list?")
my_list.reverse()
print("Answer:", my_list)

# Question: How do you sort the elements of the list in ascending order?
print("Question: How do you sort the elements of the list in ascending order?")
my_list.sort()
print("Answer:", my_list)

# Question: How do you remove and return the last element from the list?
print("Question: How do you remove and return the last element from the list?")
last_element = my_list.pop()
print("Answer:", last_element)
print("List after pop operation:", my_list)

# Question: How do you slice a portion of the list?
print("Question: How do you slice a portion of the list?")
sliced_portion = my_list[1:3]  # Slice from index 1 to index 2 (not inclusive)
print("Answer:", sliced_portion)

# Question: How do you check if a list is empty?
print("Question: How do you check if a list is empty?")
empty_list = []
print("Answer:", len(empty_list) == 0)

# Question: How do you concatenate two lists?
print("Question: How do you concatenate two lists?")
new_list = my_list + [6, 7, 8]
print("Answer:", new_list)
```

```python
# Define a tuple
my_tuple = (1, 2, 3, 4, 5)

# Question: How do you access the first element of the tuple?
print("Question: How do you access the first element of the tuple?")
print("Answer:", my_tuple[0])

# Question: How do you access the last element of the tuple?
print("Question: How do you access the last element of the tuple?")
print("Answer:", my_tuple[-1])

# Question: Can you change the elements of a tuple once it's created?
print("Question: Can you change the elements of a tuple once it's created?")
print("Answer: No, tuples are immutable, meaning their elements cannot be changed after creation.")

# Question: How do you concatenate two tuples?
print("Question: How do you concatenate two tuples?")
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
concatenated_tuple = tuple1 + tuple2
print("Answer:", concatenated_tuple)

# Question: How do you find the index of a specific element in the tuple?
print("Question: How do you find the index of a specific element in the tuple?")
print("Answer:", my_tuple.index(3))

# Question: How do you count the occurrences of a specific element in the tuple?
print("Question: How do you count the occurrences of a specific element in the tuple?")
print("Answer:", my_tuple.count(3))

# Question: Can you convert a tuple to a list?
print("Question: Can you convert a tuple to a list?")
converted_list = list(my_tuple)
print("Answer:", converted_list)

# Question: Can you convert a list to a tuple?
print("Question: Can you convert a list to a tuple?")
list_to_convert = [1, 2, 3, 4, 5]
converted_tuple = tuple(list_to_convert)
print("Answer:", converted_tuple)
```

```python
# Define a dictionary
my_dict = {"apple": 10, "banana": 5, "orange": 8, "grape": 12, "kiwi": 3}

# Question: How do you access the value associated with the key 'banana'?
print("Question: How do you access the value associated with the key 'banana'?")
print("Answer:", my_dict["banana"])

# Question: How do you add a new key-value pair to the dictionary?
print("Question: How do you add a new key-value pair to the dictionary?")
my_dict["melon"] = 7
print("Answer:", my_dict)

# Question: How do you remove a key-value pair from the dictionary?
print("Question: How do you remove a key-value pair from the dictionary?")
del my_dict["orange"]
print("Answer:", my_dict)

# Question: How do you check if a key exists in the dictionary?
print("Question: How do you check if a key exists in the dictionary?")
print("Answer:", "grape" in my_dict)

# Question: How do you find the number of key-value pairs in the dictionary?
print("Question: How do you find the number of key-value pairs in the dictionary?")
print("Answer:", len(my_dict))
```
