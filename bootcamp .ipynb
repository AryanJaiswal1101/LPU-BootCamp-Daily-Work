```python
# Creating a list of integers
numbers = [1, 2, 3, 4, 5]

# Creating a list of strings
fruits = ["apple", "banana", "cherry"]

# Creating a mixed list
mixed = [1, "apple", 3.14, True]

print(numbers)  # Output: [1, 2, 3, 4, 5]
print(fruits)   # Output: ["apple", "banana", "cherry"]
print(mixed)    # Output: [1, "apple", 3.14, True]
```

```python
fruits = ["apple", "banana", "cherry"]

# Accessing the first element
print(fruits[0])  # Output: apple

# Accessing the second element
print(fruits[1])  # Output: banana

# Accessing the last element
print(fruits[-1]) # Output: cherry
```

```python
fruits = ["apple", "banana"]

# Adding an element to the end of the list
fruits.append("cherry")
print(fruits)  # Output: ["apple", "banana", "cherry"]

# Inserting an element at a specific position
fruits.insert(1, "orange")
print(fruits)  # Output: ["apple", "orange", "banana", "cherry"]

# Removing a specific element
fruits.remove("orange")
print(fruits)  # Output: ["apple", "banana", "cherry"]

# Removing the last element
last_fruit = fruits.pop()
print(last_fruit)  # Output: cherry
print(fruits)      # Output: ["apple", "banana"]

# Deleting an element by index
del fruits[0]
print(fruits)  # Output: ["banana"]
```

```python
numbers = [1, 2, 3, 4, 5]

# Slicing from index 1 to 3 (exclusive)
print(numbers[1:4])  # Output: [2, 3, 4]

# Slicing from the beginning to index 3 (exclusive)
print(numbers[:4])   # Output: [1, 2, 3, 4]

# Slicing from index 2 to the end
print(numbers[2:])   # Output: [3, 4, 5]

# Slicing the entire list with a step
print(numbers[::2])  # Output: [1, 3, 5]
```

```python
fruits = ["apple", "banana", "cherry"]

# Iterating over the list
for fruit in fruits:
    print(fruit)

# Output:
# apple
# banana
# cherry

# Iterating with index using enumerate()
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# Output:
# Index 0: apple
# Index 1: banana
# Index 2: cherry
```

```python
# Creating a set using curly braces
my_set = {1, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}

# Creating an empty set using set() function
empty_set = set()
print(empty_set)  # Output: set()
```

```python
# Creating a set
set1 = {1, 2, 3}

# Adding an element
set1.add(4)
print(set1)  # Output: {1, 2, 3, 4}

# Removing an element using remove() (raises KeyError if the element is not found)
set1.remove(2)
print(set1)  # Output: {1, 3, 4}

# Removing an element using discard() (does not raise an error if the element is not found)
set1.discard(3)
print(set1)  # Output: {1, 4}

# Trying to remove an element that is not in the set using remove()
# set1.remove(5)  # Raises KeyError

# Trying to remove an element that is not in the set using discard()
set1.discard(5)  # Does nothing
print(set1)  # Output: {1, 4}
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union
union_set = set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)  # Output: {3, 4}

# Difference
difference_set = set1.difference(set2)
print(difference_set)  # Output: {1, 2}

# Using operators for the same operations
print(set1 | set2)  # Output: {1, 2, 3, 4, 5, 6} (Union)
print(set1 & set2)  # Output: {3, 4} (Intersection)
print(set1 - set2)  # Output: {1, 2} (Difference)
```

```python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

# Checking if set1 is a subset of set2
print(set1.issubset(set2))  # Output: True

# Checking if set2 is a superset of set1
print(set2.issuperset(set1))  # Output: True

# Using operators for the same checks
print(set1 <= set2)  # Output: True (subset check)
print(set2 >= set1)  # Output: True (superset check)
```

```python
# Creating a set
set1 = {1, 2, 3, 4, 5}

# Iterating through the set
for element in set1:
    print(element)

# Output:
# 1
# 2
# 3
# 4
# 5
```

```python
# Creating a dictionary using curly braces
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Creating a dictionary using the dict() function
person_dict = dict(name="John", age=30, city="New York")

# Example usage
print(person)       # Output: {'name': 'John', 'age': 30, 'city': 'New York'}
print(person_dict)  # Output: {'name': 'John', 'age': 30, 'city': 'New York'}
```

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Accessing values
print(person["name"])  # Output: John
print(person.get("age"))  # Output: 30

# Modifying values
person["age"] = 31
print(person["age"])  # Output: 31
```

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Adding a key-value pair
person["email"] = "john@example.com"
print(person)  # Output: {'name': 'John', 'age': 30, 'city': 'New York', 'email': 'john@example.com'}

# Removing a key-value pair using del
del person["city"]
print(person)  # Output: {'name': 'John', 'age': 30, 'email': 'john@example.com'}

# Removing a key-value pair using pop()
age = person.pop("age")
print(age)     # Output: 30
print(person)  # Output: {'name': 'John', 'email': 'john@example.com'}
```

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Iterating over keys
for key in person.keys():
    print(key)  # Output: name, age, city

# Iterating over values
for value in person.values():
    print(value)  # Output: John, 30, New York

# Iterating over key-value pairs
for key, value in person.items():
    print(f"{key}: {value}")
    # Output: 
    # name: John
    # age: 30
    # city: New York
```

```python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Checking if a key exists
if "name" in person:
    print("Name exists in the dictionary")  # Output: Name exists in the dictionary

if "email" not in person:
    print("Email does not exist in the dictionary")  # Output: Email does not exist in the dictionary
```
