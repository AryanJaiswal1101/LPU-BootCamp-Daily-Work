```python
import tkinter as tk
from tkinter import messagebox

# List of questions and answers
qa_list = [
    {"question": "What is the capital of France?", "answer": "Paris"},
    {"question": "What is 2 + 2?", "answer": "4"},
    {"question": "What is the capital of Japan?", "answer": "Tokyo"},
]

# Index of the current question
current_index = 0

# Function to check the answer
def check_answer():
    user_answer = answer_entry.get()
    correct_answer = qa_list[current_index]['answer']
    if user_answer.lower() == correct_answer.lower():
        result_label.config(text="Correct!", fg="green")
    else:
        result_label.config(text="Incorrect. Try again.", fg="red")

# Function to show the next question
def next_question():
    global current_index
    if current_index < len(qa_list) - 1:
        current_index += 1
        display_question()
    else:
        messagebox.showinfo("End", "This is the last question.")

# Function to show the previous question
def prev_question():
    global current_index
    if current_index > 0:
        current_index -= 1
        display_question()
    else:
        messagebox.showinfo("Start", "This is the first question.")

# Function to display the current question
def display_question():
    question_label.config(text=qa_list[current_index]['question'])
    answer_entry.delete(0, tk.END)
    result_label.config(text="")

# Create the main window
root = tk.Tk()
root.title("Question and Answer")

# Create a label to display the question
question_label = tk.Label(root, text="")
question_label.pack(pady=10)

# Create an entry widget for the user to type their answer
answer_entry = tk.Entry(root)
answer_entry.pack(pady=10)

# Create a button to submit the answer
submit_button = tk.Button(root, text="Submit", command=check_answer)
submit_button.pack(pady=10)

# Create a label to display the result
result_label = tk.Label(root, text="")
result_label.pack(pady=10)

# Create buttons to navigate between questions
prev_button = tk.Button(root, text="Previous", command=prev_question)
prev_button.pack(side=tk.LEFT, padx=20)

next_button = tk.Button(root, text="Next", command=next_question)
next_button.pack(side=tk.RIGHT, padx=20)

# Display the first question
display_question()

# Run the application
root.mainloop()
```
